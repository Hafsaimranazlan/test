```python
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies.genres.apply(convert)
```

```python
 movies.head(1)
```

```python
 movies.keywords=movies.keywords.apply(convert)
```

```python
def convert2(obj):
    l=[]
    count=0
    for i in ast.literal_eval(obj):
        if count != 3:
            l.append(i['name'])
            count+=1
        else:
            break
    return l
```

```python
movies['cast']=movies.cast.apply(convert2)
```

```python
movies.head(1)
```

```python
#movies.crew[0]
```

```python
def convert3(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies.crew=movies.crew.apply(convert3)
```

```python
 movies.head(2)
```

```python
movies.overview=movies.overview.apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
movies['cast']=movies.cast.apply(lambda x:[i.replace(" ","") for i in x]) 
movies['crew']=movies.crew.apply(lambda x:[i.replace(" ","") for i in x]) 
movies['genres']=movies.genres.apply(lambda x:[i.replace(" ","") for i in x]) 
movies['keywords']=movies.keywords.apply(lambda x:[i.replace(" ","") for i in x]) 
```

```python
movies.head(2)
```

```python
movies['tags']=movies.overview + movies.keywords + movies.genres + movies.cast + movies.crew
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df.tags=new_df.tags.apply(lambda x:" ".join(x))
```

```python
new_df.tags[0]
```

```python
new_df.head(2)
```

```python
new_df['tags']=new_df.tags.apply(lambda x:x.lower())
```

```python
new_df.head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=5000,stop_words='english')

vector=cv.fit_transform(new_df['tags']).toarray()

sorted(cv.vocabulary_)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
new_df.tags=new_df.tags.apply(stem)
```

```python
new_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
#sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]  
```

```python
def recomend(movie):
    new_index=new_df[new_df['title']==movie].index[0]
    dist=similarity[new_index]
    recomend_movie=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6] 
    for i in recomend_movie:
        print(new_df.iloc[i[0]].title)
```

```python
recomend('Fear Clinic')
```
